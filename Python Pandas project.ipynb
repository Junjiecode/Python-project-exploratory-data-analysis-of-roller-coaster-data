# Step 2 Data preparation

```python
# Rename the columns : capitalize the 1st letter 
df = df.rename(columns = {"latitude": "Latitude", "longitude":"Longitude", "speed_mph":"Speed_mph", \
                         "year_introduced" : "Year_introduced", "height_ft":"Height_ft"})
```

```python
# Convert the "Duration" to unique data format 

# step 1: define Define the Conversion Function
import pandas as pd
import numpy as np
import re

def convert_to_seconds_simple(duration):
    if pd.isnull(duration):
        return np.nan

    # Define regex patterns for different formats
    minute_second_pattern = re.compile(r'(\d+):(\d+)')
    minute_only_pattern = re.compile(r'(\d+)\s*minute')
    second_only_pattern = re.compile(r'(\d+)\s*second')
    generic_pattern = re.compile(r'(\d+)\s*min|(\d+)\s*sec')

    # Handle "MM:SS" format
    match = minute_second_pattern.match(duration)
    if match:
        minutes, seconds = map(int, match.groups())
        return minutes * 60 + seconds

    # Handle "X minute Y seconds" format
    match = re.match(r'(\d+)\s*minute\s*(\d+)\s*seconds?', duration)
    if match:
        minutes, seconds = map(int, match.groups())
        return minutes * 60 + seconds

    # Handle "X minutes" format
    match = minute_only_pattern.match(duration)
    if match:
        minutes = int(match.group(1))
        return minutes * 60

    # Handle "X seconds" format
    match = second_only_pattern.match(duration)
    if match:
        seconds = int(match.group(1))
        return seconds

    # Handle generic "X min" or "X sec" format
    match = generic_pattern.match(duration)
    if match:
        minutes = int(match.group(1)) if match.group(1) else 0
        seconds = int(match.group(2)) if match.group(2) else 0
        return minutes * 60 + seconds

    return np.nan

# step 2: apply the function
df["Duration"] = df["Duration"].apply(convert_to_seconds_simple)
```

```python
# What're the location with the fastest roller coasters? (with minimun of 10)

df["Location"].value_counts()

df.query('Location != "Other"').\
groupby("Location")["Speed_mph"].\
agg(["count", "mean"]).\
query('count >= 10').\
sort_values("mean")["mean"].\
plot(kind = 'barh', figsize = (12,5)).\
set_title("Top 10  AVG coaster speed by Location ")


plt.show()
```

```python
# When was the first roller coasters built in the records, and where it was?

df.sort_values("Year_introduced")[["Year_introduced", "Location"]].head(1)
```

```python
# Who are the top 10 manufacturer who constract the most of roller coasters?

df.groupby("Manufacturer")["coaster_name"]\
    .count()\
    .sort_values(ascending=False)\
    .head(10)\
    .plot(kind="barh")
```
